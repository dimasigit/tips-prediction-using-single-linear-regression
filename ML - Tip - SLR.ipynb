```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
import statsmodels
import warnings
warnings.filterwarnings('ignore')
from sklearn import linear_model
```

```python
df_tips = sns.load_dataset('tips')
```

```python
df_tips.head()
```

```python
x= df_tips[['total_bill']]
y=df_tips[['tip']]
```

```python
Model = linear_model.LinearRegression()
```

```python
Model.fit(x,y)
```

```python
df_tips['Prediksi'] = Model.predict(x).round(2)
df_tips
# membuat kolom 'prediksi' yg isinya prediksi x
```

```python
plt.scatter(df_tips['total_bill'], df_tips['tip'], color='blue')
plt.plot(df_tips['total_bill'], df_tips['Prediksi'], color='red')
plt.show()
```

```python
df_tips['tip_error'] = abs(df_tips['tip']-df_tips['Prediksi'])
df_tips
# membuat kolom tip error
```

```python
sns.distplot(df_tips['tip_error'])
plt.show()
```

```python
sns.boxplot(df_tips['tip_error'])
plt.show()
```
